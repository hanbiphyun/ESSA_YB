##12-1. datetime 오브젝트

: datatime 라이브러리는 날짜와 시간을 처리하는 등의 다양한 기능을 제공하는 파이썬 라이브러리

-> date(날짜), time(시간), datetime(날짜+시간) 오브젝트 포함

- to_datetime 메서드 : 시간 데이터를 문자열로 저장할때, 문자열을 datetime 오브젝트로 변환
> 시간 형식 지정자(%d, %m, %y)와 기호
(/, -)를 적절히 조합하여 format 인자에 전달하면 그 형식에 맞게 정리된 datetime 오브젝트를 얻을 수 있다.

```python
#데이터 프레임 준비하기
ebola= pd.read_csv('/content/country_timeseries.csv', parse_dates=['Date'] )
print(ebola.head().iloc[:,:5])
print(ebola.tail().iloc[:,:5])
```

```python
ebola.index= ebola['Date'] #Date 열을 인덱스로 지정
new_idx= pd.date_range(ebola.index.min(), ebola.index.max()) #최댓값과 최솟값으로 시간 범위 생성
print(new_idx)
#시간이 역순으로 나옴

new_idx=reversed(new_idx)
print(new_idx)
```

```python
#reindex 메서드 사용-> 새로 생성한 인덱스를 새 인덱스로 지정 -> 없던 날짜 추가
ebola=ebola.reindex(new_idx)
print(ebola.head().iloc[:,:5])
```

```python
print(ebola.tail().iloc[:,:5])
```

```python
#last_valid_index, first_valid_index 메서드 이용-> 발병일 구하기
last_valid= ebola.apply(pd.Series.last_valid_index) #가장 오래된 데이터
print(last_valid)
```

```python
first_valid= ebola.apply(pd.Series.first_valid_index)
print(first_valid)
```

```python
#발병일을 동일한 출발선으로 옮기기
#처음 발병한 날 - 각 나라 발병일 만큼 옮기기

earliest_date= ebola.index.min()
print(earliest_date)
```

```python
shift_values= last_valid- earliest_date
print(shift_values)
```

```python
#shift 메서드 : 인잣값만큼 데이터 밀어냄
ebola_dict={}
for idx, col in enumerate(ebola):
    d= shift_values[idx].days
    shifted= ebola[col].shift(d)
    ebola_dict[col]= shifted
```

```python
ebola_shift= pd.DataFrame(ebola_dict) #딕셔너리 형태로 저장된 것 -> 데이터프레임으로 변환
```

```python
print(ebola_shift.tail())
```

```python
#인덱스 Day열로 지정, 필요없는 Date, Day열 삭제
ebola_shift.index= ebola_shift['Day']
ebola_shift= ebola_shift.drop(['Date','Day'], axis=1)
print(ebola_shift.tail())
```

```python
fig, ax= plt.subplots()
ax= ebola_shift.iloc[:,:].plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
```
