```python
# 핵심코드
import torch
import torch.nn as nn
from transformers import AutoModel, AutoConfig

class CommonLitModel(nn.Module):
    def __init__(self, model_name):
        super(CommonLitModel, self).__init__()
        # 1. 사전 학습된 모델의 설정(Config) 로드
        self.config = AutoConfig.from_pretrained(model_name)
        # 2. Transformer 백본 모델 (예: RoBERTa, DeBERTa)
        self.model = AutoModel.from_pretrained(model_name, config=self.config)

        # 3. 우승 팀의 핵심: 단순 CLS 토큰이 아닌 Attention Pooling 사용
        self.attention = nn.Sequential(
            nn.Linear(self.config.hidden_size, 512),
            nn.Tanh(),
            nn.Linear(512, 1),
            nn.Softmax(dim=1)
        )

        # 4. 최종 점수(난이도)를 출력하는 Linear 레이어
        self.regressor = nn.Linear(self.config.hidden_size, 1)

    def forward(self, input_ids, attention_mask):
        # Transformer 통과
        outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
        last_hidden_state = outputs.last_hidden_state # [batch_size, seq_len, hidden_size]

        # Attention Weights 계산
        weights = self.attention(last_hidden_state)

        # 문장의 각 단어 벡터를 가중 평균하여 하나의 문장 벡터 생성 (Context Vector)
        context_vector = torch.sum(weights * last_hidden_state, dim=1)

        # 최종 난이도 점수 예측
        return self.regressor(context_vector)

# 모델 생성 예시 (RoBERTa-Large 사용)
model = CommonLitModel('roberta-large')
```
