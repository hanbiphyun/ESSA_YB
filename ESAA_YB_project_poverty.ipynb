##결측치

```python
# 결측치 비율 계산
missing_info = train.isnull().mean() * 100

# 컬럼 리스트를 10개씩 나누기
cols = missing_info.index.tolist()
for i in range(0, len(cols), 10):
    print(f"--- {i}번부터 {i+9}번 컬럼 결측치 (%) ---")
    print(missing_info.iloc[i:i+10])
    print("-" * 30)
    # input("다음 10개를 보려면 Enter를 누르세요...") # 주피터 환경에서 하나씩 끊어보고 싶을 때 사용
```

```python
# TOP 5 일괄 처리 코드
# 1. sector1d
train['sector1d'] = train['sector1d'].fillna(-1)
test_features['sector1d'] = test_features['sector1d'].fillna(-1)

# 2. dweltyp
mode_val = train['dweltyp'].mode()[0]
train['dweltyp'] = train['dweltyp'].fillna(mode_val)
test_features['dweltyp'] = test_features['dweltyp'].fillna(mode_val)

# 3. utl_exp_ppp17
median_val = train['utl_exp_ppp17'].median()
train['utl_exp_ppp17'] = train['utl_exp_ppp17'].fillna(median_val)
test_features['utl_exp_ppp17'] = test_features['utl_exp_ppp17'].fillna(median_val)
```

```python
emp_mode = train['employed'].mode()[0]
train['employed'] = train['employed'].fillna(emp_mode)
test_features['employed'] = test_features['employed'].fillna(emp_mode)
```

```python
edu_mode = train['educ_max'].mode()[0]
train['educ_max'] = train['educ_max'].fillna(edu_mode)
test_features['educ_max'] = test_features['educ_max'].fillna(edu_mode)
```

```python
edu_median = train['share_secondary'].median()
train['share_secondary'] = train['share_secondary'].fillna(edu_median)
test_features['share_secondary'] = test_features['share_secondary'].fillna(edu_median)
```

```python
# 전체 데이터에서 결측치가 단 1개라도 남아있는 컬럼 출력
final_missing = train.isnull().sum()
print("--- 아직도 결측치가 남아있는 컬럼 ---")
print(final_missing[final_missing > 0])
```

```python
import pandas as pd
import numpy as np

# 1. 매핑 사전 정의 (텍스트 -> 숫자)
# 데이터에 있는 다양한 텍스트 형식을 숫자로 바꿉니다.
mapping = {
    'Access': 1, 'No access': 0,
    'Yes': 1, 'No': 0,
    'Urban': 1, 'Rural': 0,
    'Male': 1, 'Female': 0
}

# 2. 변환할 컬럼 리스트
infra_cols = ['elect', 'water', 'toilet', 'sewer']
other_cat_cols = ['male', 'owner', 'urban', 'any_nonagric']
consumed_cols = [c for c in train.columns if 'consumed' in c]

# 3. 일괄 변환 함수
def convert_text_to_num(df):
    target_cols = infra_cols + other_cat_cols + consumed_cols
    for col in target_cols:
        if col in df.columns and df[col].dtype == 'object':
            # 매핑 사전에 있는 값은 바꾸고, 없는 값은 그대로 둡니다.
            df[col] = df[col].map(mapping).fillna(df[col])
            # 혹시나 남아있을 수 있는 결측치는 0으로 채우고 숫자로 변환
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    return df

# Train과 Test에 적용
train = convert_text_to_num(train)
test_features = convert_text_to_num(test_features)

print("변환 완료! 이제 모든 인프라 변수가 0과 1로 바뀌었습니다.")
```

```python
from sklearn.preprocessing import LabelEncoder

# 1. 아직 object(문자열) 타입인 컬럼들만 추출
remaining_obj_cols = train.select_dtypes(include=['object']).columns
print("남은 문자열 컬럼들:", remaining_obj_cols.tolist())

# 2. LabelEncoder로 일괄 변환
le = LabelEncoder()

for col in remaining_obj_cols:
    # Train과 Test에 있는 모든 고유값을 합쳐서 학습 (에러 방지)
    # 결측치가 있다면 'Unknown'으로 임시 처리
    train[col] = train[col].astype(str).fillna('Unknown')
    if col in test_features.columns:
        test_features[col] = test_features[col].astype(str).fillna('Unknown')

    # 레이블 인코딩 적용
    le.fit(list(train[col]) + list(test_features[col]))
    train[col] = le.transform(train[col])
    if col in test_features.columns:
        test_features[col] = le.transform(test_features[col])

print("\n모든 컬럼이 숫자로 변환되었습니다!")
print(train[remaining_obj_cols].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. 상관계수 계산
correlations = train.select_dtypes(include=[np.number]).corr()['cons_ppp17']

# 2. 양의 상관관계 상위 10개 (본인 제외)
top_pos = correlations.sort_values(ascending=False).iloc[1:11].index.tolist()

# 3. 음의 상관관계 상위 10개 (가장 낮은 값들)
top_neg = correlations.sort_values(ascending=True).iloc[:10].index.tolist()

# 4. 히트맵 대상 변수 선정 (타겟 포함)
heatmap_cols = ['cons_ppp17'] + top_pos + top_neg

# 5. 상관계수 행렬 생성
corr_matrix_sub = train[heatmap_cols].corr()

# 6. 히트맵 시각화
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix_sub,
            annot=True,            # 숫자 표시
            cmap='RdBu_r',         # 빨간색(양), 파란색(음) 색상
            fmt=".2f",             # 소수점 둘째자리까지
            center=0,              # 0을 기준으로 색상 대비
            linewidths=0.5)

plt.title('Correlation Heatmap: Top 20 Features with Consumption', fontsize=16)
plt.show()

# 음의 상관관계 리스트 확인용 출력
print("\n--- 소비액과 음의 상관관계가 높은 Top 10 변수 ---")
print(correlations.sort_values(ascending=True).head(10))
```
