Hull Starter Notebook 리뷰 요약
링크 :https://www.kaggle.com/code/laurentlanteigne/hull-starter-notebook

1. 배경

Hull 데이터셋은 예측 문제를 다루는 Kaggle 경진대회용 데이터로, 주어진 피처를 기반으로 타깃 값을 예측하는 회귀 문제임.

데이터 전처리, 모델링, 성능 평가를 중심으로 실용적인 예제 제공.

2. 데이터 분석 전략

결측치 확인 및 기본 통계 분석으로 데이터 상태를 파악함.

피처 간 상관관계를 시각화하여 타깃과 연관성이 높은 변수를 확인함.

타깃 변수 분포를 확인하여 예측 모델 선택 시 필요한 스케일링이나 변환 여부 판단.

3. 모델링 접근

단순 선형회귀 기반 모델(Ridge, Lasso)과 비선형 앙상블 모델(RandomForest)을 사용.

Ridge/Lasso는 규제를 통해 과적합 방지와 피처 선택 효과를 확인.

RandomForest는 비선형 관계를 포착하고 예측 정확도를 높이는 데 유리함.

각 모델별 RMSE를 비교하여 성능 평가.

4. 주요 특징 및 인사이트

단순 선형 모델과 비교했을 때, RandomForest가 더 나은 예측 성능을 보임.

피처 중요도 분석을 통해 타깃에 영향을 많이 미치는 변수 확인 가능.

데이터 전처리와 모델 파이프라인 구성이 명확하여 재현성이 높음.

시각화를 통해 데이터 분포, 상관관계, 이상치 여부를 직관적으로 확인 가능.

5. 보완 및 개선점

하이퍼파라미터 튜닝 부족: 모델 성능 향상을 위해 GridSearchCV나 RandomizedSearchCV 활용 가능.

앙상블 기법 추가 가능: 여러 모델 결합을 통해 예측 안정성 향상 가능.

특성 공학 부족: 피처 생성, 로그 변환, 비선형 변환 등 추가 시 모델 성능 개선 가능.

데이터 이해 부족 시, 도메인 지식을 활용한 변수 해석이 어려움.

6. 결론

모델 성능 향상과 인사이트 도출을 위해 하이퍼파라미터 튜닝, 앙상블, 특성 공학을 추가로 고려할 수 있음.
