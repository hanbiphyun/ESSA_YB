###HMS - Harmful Brain Activity Classification
**주제** : 뇌파(EEG)데이터를 이용한 발작(seizure) 및 유해한 뇌활동 분류

**목적** : 중환자, 신경내과 환자의 뇌파 모니터링에서 이상 신호를 자동으로 탐지하여 진단과 치료를 빠르게 할 수 있도록 도움

**데이터** 1. raw EEG 신호 : 여러 전극 채널에서 일정 시간동안 측정된 시계열 데이터

2. Spectrogram 이미지 : EEG신호를 시간-주파수 영역으로 변환한 이미지 형태의 데이터

3. 평가지표 : KL-Divergence(KLD)라는 지표사용, 예측분포와 실제 분포간의 차이 최소화

**특징**
- EEG는 비정상정이 강하고 노이즈가 많다. 또한 채널, 시간길이, 환자 간 차이가 크다. -> 전처리, 모델 설계가 중요하다.

###데이터 전처리
- EEG 신로 분할 후 노이즈 제거 및 신호 정규화 수행
- 각 segment에 대해 스펙트로그램 변환으로 CNN에 입력 가능하도록 이미지 형태로 만든다.

###피처엔지니어링
- 시간영역 + 주파수 영역을 함께 모델에 제공

###모델 구조
- CNN기반 모델 사용

###앙상블
- 여러 모델의 확률 출력을 가중 평균으로 구함
- log-loss 최소화

###차별화된 점 및 배울 점
- 다양한 스펙트로그램 입력을 사용하는 멀티 인풋 구조
- GroupKFold 적용으로 데이터 누수 방지
- 모델 앙상블
- 도메인 지식을 반영한 신호 처리 접근이 중요하다.
- 데이터 누수 방지 전략을 배웠다.
- 각 모델의 출력을 확률로 관리하고 log-loss를 최소화하는 전략을 사용했다.
