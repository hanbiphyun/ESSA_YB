##결측치

```python
# 결측치 비율 계산
missing_info = train.isnull().mean() * 100

# 컬럼 리스트를 10개씩 나누기
cols = missing_info.index.tolist()
for i in range(0, len(cols), 10):
    print(f"--- {i}번부터 {i+9}번 컬럼 결측치 (%) ---")
    print(missing_info.iloc[i:i+10])
    print("-" * 30)
    # input("다음 10개를 보려면 Enter를 누르세요...") # 주피터 환경에서 하나씩 끊어보고 싶을 때 사용
```

```python
# TOP 5 일괄 처리 코드
# 1. sector1d
train['sector1d'] = train['sector1d'].fillna(-1)
test_features['sector1d'] = test_features['sector1d'].fillna(-1)

# 2. dweltyp
mode_val = train['dweltyp'].mode()[0]
train['dweltyp'] = train['dweltyp'].fillna(mode_val)
test_features['dweltyp'] = test_features['dweltyp'].fillna(mode_val)

# 3. utl_exp_ppp17
median_val = train['utl_exp_ppp17'].median()
train['utl_exp_ppp17'] = train['utl_exp_ppp17'].fillna(median_val)
test_features['utl_exp_ppp17'] = test_features['utl_exp_ppp17'].fillna(median_val)
```

```python
emp_mode = train['employed'].mode()[0]
train['employed'] = train['employed'].fillna(emp_mode)
test_features['employed'] = test_features['employed'].fillna(emp_mode)
```

```python
edu_mode = train['educ_max'].mode()[0]
train['educ_max'] = train['educ_max'].fillna(edu_mode)
test_features['educ_max'] = test_features['educ_max'].fillna(edu_mode)
```

```python
edu_median = train['share_secondary'].median()
train['share_secondary'] = train['share_secondary'].fillna(edu_median)
test_features['share_secondary'] = test_features['share_secondary'].fillna(edu_median)
```

```python
# 전체 데이터에서 결측치가 단 1개라도 남아있는 컬럼 출력
final_missing = train.isnull().sum()
print("--- 아직도 결측치가 남아있는 컬럼 ---")
print(final_missing[final_missing > 0])
```

```python
import pandas as pd
import numpy as np

# 1. 매핑 사전 정의 (텍스트 -> 숫자)
# 데이터에 있는 다양한 텍스트 형식을 숫자로 바꿉니다.
mapping = {
    'Access': 1, 'No access': 0,
    'Yes': 1, 'No': 0,
    'Urban': 1, 'Rural': 0,
    'Male': 1, 'Female': 0
}

# 2. 변환할 컬럼 리스트
infra_cols = ['elect', 'water', 'toilet', 'sewer']
other_cat_cols = ['male', 'owner', 'urban', 'any_nonagric']
consumed_cols = [c for c in train.columns if 'consumed' in c]

# 3. 일괄 변환 함수
def convert_text_to_num(df):
    target_cols = infra_cols + other_cat_cols + consumed_cols
    for col in target_cols:
        if col in df.columns and df[col].dtype == 'object':
            # 매핑 사전에 있는 값은 바꾸고, 없는 값은 그대로 둡니다.
            df[col] = df[col].map(mapping).fillna(df[col])
            # 혹시나 남아있을 수 있는 결측치는 0으로 채우고 숫자로 변환
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    return df

# Train과 Test에 적용
train = convert_text_to_num(train)
test_features = convert_text_to_num(test_features)

print("변환 완료! 이제 모든 인프라 변수가 0과 1로 바뀌었습니다.")
```

```python
from sklearn.preprocessing import LabelEncoder

# 1. 아직 object(문자열) 타입인 컬럼들만 추출
remaining_obj_cols = train.select_dtypes(include=['object']).columns
print("남은 문자열 컬럼들:", remaining_obj_cols.tolist())

# 2. LabelEncoder로 일괄 변환
le = LabelEncoder()

for col in remaining_obj_cols:
    # Train과 Test에 있는 모든 고유값을 합쳐서 학습 (에러 방지)
    # 결측치가 있다면 'Unknown'으로 임시 처리
    train[col] = train[col].astype(str).fillna('Unknown')
    if col in test_features.columns:
        test_features[col] = test_features[col].astype(str).fillna('Unknown')

    # 레이블 인코딩 적용
    le.fit(list(train[col]) + list(test_features[col]))
    train[col] = le.transform(train[col])
    if col in test_features.columns:
        test_features[col] = le.transform(test_features[col])

print("\n모든 컬럼이 숫자로 변환되었습니다!")
print(train[remaining_obj_cols].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# 1. 분석 결과에 따른 불필요한 변수 제거
# 'owner'는 모든 값이 0이므로 변별력이 없음
drop_cols = ['owner']
train.drop(columns=[col for col in drop_cols if col in train.columns], inplace=True)
test_features.drop(columns=[col for col in drop_cols if col in test_features.columns], inplace=True)

# 2. 피처 엔지니어링 (가설 기반 파생 변수 생성)
def engineer_features(df):
    # (1) 주거 인프라 종합 점수 (가설 1 기반)
    infra_list = ['elect', 'water', 'toilet', 'sewer']
    df['infra_score'] = df[infra_list].sum(axis=1)

    # (2) 사치재 소비 점수 (가설 3 기반 - 소고기, 버터, 외식 등 하위 소비 10개 품목)
    # 분석에서 도출된 하위 10개 품목 번호를 활용
    luxury_items = ['consumed2600', 'consumed2200', 'consumed1300', 'consumed4400', 'consumed5000',
                    'consumed1400', 'consumed1200', 'consumed4300', 'consumed4900', 'consumed2500']
    df['luxury_food_score'] = df[luxury_items].sum(axis=1)

    # (3) 식생활 다양성 점수 (전체 소비 품목 합계)
    consumed_cols = [c for c in df.columns if 'consumed' in c and c not in luxury_items]
    df['total_diet_diversity'] = df[consumed_cols].sum(axis=1)

    # (4) 가구 부양 부담 지표 (가구원 수 대비 성인 비율 등)
    df['adult_ratio'] = (df['num_adult_female'] + df['num_adult_male']) / df['hsize']

    return df

train = engineer_features(train)
test_features = engineer_features(test_features)

# 3. 범주형 변수 처리 (Label Encoding)
# 이미 숫자로 변환되지 않은 나머지 object 컬럼들을 처리합니다.
le = LabelEncoder()
obj_cols = train.select_dtypes(include=['object']).columns

for col in obj_cols:
    if col in ['survey_id', 'sample']: continue # 식별용 제외

    # Train/Test 통합 학습으로 카테고리 불일치 방지
    combined_data = pd.concat([train[col].astype(str), test_features[col].astype(str)], axis=0)
    le.fit(combined_data)

    train[col] = le.transform(train[col].astype(str))
    test_features[col] = le.transform(test_features[col].astype(str))

# 4. 타겟 변수 로그 변환 (EDA 결과 반영)
# 학습 직전에 적용하기 위해 원본을 따로 보관하거나, y값만 따로 추출할 때 적용합니다.
train['log_cons_ppp17'] = np.log1p(train['cons_ppp17'])

print("최종 전처리 및 파생 변수 생성이 완료되었습니다.")
print(f"현재 변수 개수: {len(train.columns)}개")
```

###region7의 부자인 경우

  - consumed4600, 4700 생수,외식 -> 물을 사먹고 외식하는 비중이 높다
  - consumed4400 주류소비도 20% 높음
2. 경제 지표
- utl_exp_ppp17 공공요금 : 부자 계층이 빈곤 계층보다 훨씬 많이 냄
- sfworkershh 정규직 비율 : 부유층의 정규직 비율이 높음, 부의 핵심인 안정적인 소득원

=> 추가적인 파생변수 생성해보자

```python
def region7_specific_engineering(df):
    # 1. Region 7 전용 '엘리트 지표'
    # Region 7에 살면서 + 버터를 먹고 + 외식을 자주하며 + 정규직인 경우
    df['r7_elite_index'] = (
        (df['region7'] == 1).astype(int) * (df['consumed2600'] + df['consumed4700'] + (df['sfworkershh'] > 0.5).astype(int))
    )

    # 2. 고액 공공요금 아웃라이어 (Region 7 내의 상대적 부유함)
    r7_utl_mean = df[df['region7'] == 1]['utl_exp_ppp17'].mean()
    df['r7_high_utility'] = ((df['region7'] == 1) & (df['utl_exp_ppp17'] > r7_utl_mean)).astype(int)

    # 3. 사치재 소비 조합 (버터 + 생수 + 주류 + 외식)
    luxury_items = ['consumed2600', 'consumed4600', 'consumed4400', 'consumed4700']
    df['luxury_lifestyle_score'] = df[luxury_items].sum(axis=1)

    return df

train = region7_specific_engineering(train)
test_features = region7_specific_engineering(test_features)
```

```python
def optimize_education_features(df):
    # 1. 박스플롯에서 확인된 '소비가 높은 특정 학력'만 추출 (One-Hot 느낌)
    # 지수 2와 5에서만 소비가 높았으므로 이 정보만 명시적으로 제공
    df['edu_high_consumption_spot'] = df['educ_max'].apply(lambda x: 1 if x in [2, 5] else 0)

    # 2. 교육과 고용의 결합 (여전히 유효한 전략)
    # 교육 단독보다는 '교육받은 정규직'이라는 맥락이 더 중요할 수 있음
    df['edu_job_synergy'] = df['educ_max'] * df['sfworkershh']

    # [저소득 지역 통합] Region 4, 5, 6 통합 (삭제하기 전에 실행)
    df['low_growth_area'] = df[['region4', 'region5', 'region6']].sum(axis=1)

    # 3. 원본 변수 및 불필요한 변수 최종 삭제
    # 상관계수가 낮았던 원본 educ_max와 이미 그룹화한 지역 변수들 삭제
    drop_cols = ['educ_max', 'owner', 'com', 'region4', 'region5', 'region6']
    df = df.drop(columns=[c for c in drop_cols if c in df.columns])

    return df

# 적용
train = optimize_education_features(train)
test_features = optimize_education_features(test_features)

print(f"✅ 교육 변수 최적화 완료! 남은 피처 수: {len(train.columns)}개")
print(f"새로 만든 'edu_high_consumption_spot'의 평균 소비액 확인:")
print(train.groupby('edu_high_consumption_spot')['cons_ppp17'].mean())
```

```python
# 가설 : 단백질을 포함하는 식사를 하는 가구의 경우 소비수준이 높을것이다.
# 1. 품목 분류
carbs_items = [100, 300, 500, 1500, 1600, 1700, 1800, 1900, 3500, 3600] # 곡물, 감자, 옥수수 등
protein_items = [400, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 2000, 2100, 2400, 2600, 2700] # 유제품, 달걀, 육류, 생선

carbs_cols = [f'consumed{i}' for i in carbs_items if f'consumed{i}' in train.columns]
protein_cols = [f'consumed{i}' for i in protein_items if f'consumed{i}' in train.columns]

# 2. 가구별 섭취 개수 계산
train['carbs_count'] = train[carbs_cols].sum(axis=1)
train['protein_count'] = train[protein_cols].sum(axis=1)

# 3. 단백질/탄수화물 비율 (식사의 질 지표)
# 분모가 0이 되는 것을 방지하기 위해 +1
train['protein_to_carbs_ratio'] = train['protein_count'] / (train['carbs_count'] + 1)

# 4. 시각화: 식단 비율과 소비액의 관계
plt.figure(figsize=(10, 6))
sns.regplot(x='protein_to_carbs_ratio', y='cons_ppp17', data=train,
            scatter_kws={'alpha':0.1}, line_kws={'color':'red'})
plt.title('Protein/Carbs Ratio vs Daily Consumption', fontsize=15)
plt.show()
```

###추가 파생변수

```python
# 파생변수 생성 및 오류 수정 버전
def advanced_feature_engineering(df):
    # 1. 의존 비율 (부양가족 수 / 경제활동 인구)
    df['dependency_ratio'] = (df['num_children18'] + df['num_elderly']) / (df['num_adult_female'] + df['num_adult_male'] + 1)

    # 2. 1인당 인프라 혜택 (가구원이 많을수록 인프라 체감이 낮을 수 있음)
    df['infra_per_person'] = df['infra_score'] / (df['hsize'] + 1)

    # 3. 고용 안정성 및 산업 특성 결합
    df['employment_quality'] = (df['sfworkershh'] * 1.5) + df['any_nonagric']

    # 4. 나이대별 생애주기 그룹화 (astype(int)를 통해 확실히 숫자로 변환)
    # pd.cut 결과가 'Senior' 같은 문자열로 남아있지 않도록 합니다.
    df['age_group'] = pd.cut(df['age'], bins=[0, 19, 39, 59, 100], labels=[0, 1, 2, 3])
    df['age_group'] = df['age_group'].astype(int)

    # 5. 도시화와 인프라의 상호작용 (도시인데 인프라가 낮으면 극빈층 확률 높음)
    df['urban_infra_mismatch'] = df['urban'] * (4 - df['infra_score'])

    # 6. 식단 다양성 정규화 (가구원 수 대비 다양성)
    df['diet_div_per_person'] = df['total_diet_diversity'] / (df['hsize'] + 1)

    # 7. 불필요한 노이즈 변수 및 EDA용 문자열 변수 제거
    # 'age_bin'이 남아있으면 corr()에서 오류가 발생하므로 여기서 함께 삭제합니다.
    cols_to_drop = [ 'com', 'age_bin']
    df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

    return df

# 적용
train = advanced_feature_engineering(train)
test_features = advanced_feature_engineering(test_features)

print(f"심화 파생 변수 생성 완료! 현재 총 변수 개수: {len(train.columns)}개")

# 상관관계 재확인 (숫자형 데이터만 골라서 계산하여 'Senior' 같은 문자열 오류 원천 차단)
numeric_train = train.select_dtypes(include=[np.number])
new_corr = numeric_train.corr()['cons_ppp17'].sort_values(ascending=False)

print("\n--- 새로 추가된 변수들의 상관계수 ---")
print(new_corr.head(10))
```

###모델링

```python
import lightgbm as lgb
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd

# 1. 피처와 타겟 설정
exclude_cols = ['survey_id', 'hhid', 'sample', 'cons_ppp17', 'weight', 'log_cons_ppp17']
features = [col for col in train.columns if col in test_features.columns and col not in exclude_cols]

# Pandas DataFrame 형태를 유지하면서 .values로 추출
X = train[features].values
y = np.log1p(train['cons_ppp17']).values.flatten() # 1차원 보장
weights = train['weight'].values.flatten()
test_X = test_features[features].values

# 2. 5-Fold 교차 검증 학습
kf = KFold(n_splits=5, shuffle=True, random_state=42)
test_preds = np.zeros(len(test_features))

print(f"학습 시작 (사용 피처 수: {len(features)}개)...")

# LightGBM 파라미터 정의
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'verbosity': -1,
    'learning_rate': 0.05,
    'boosting_type': 'gbdt'
}

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    # 인덱스를 사용하여 데이터 분할 (2차원 유지)
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]
    w_train = weights[train_idx]

    # Dataset 생성 시 데이터 형태 강제 확인
    train_set = lgb.Dataset(X_train, label=y_train, weight=w_train, free_raw_data=False)
    val_set = lgb.Dataset(X_val, label=y_val, reference=train_set, free_raw_data=False)

    # 모델 학습 (valid_sets를 추가하여 안정성 높임)
    model = lgb.train(
        params,
        train_set,
        num_boost_round=1000,
        valid_sets=[val_set],
        callbacks=[lgb.early_stopping(stopping_rounds=100)]
    )

    # 테스트 데이터 예측
    test_preds += model.predict(test_X, num_iteration=model.best_iteration) / 5

# 결과 복구
final_predictions = np.expm1(test_preds)
print("\n✅ 모델 학습 및 예측이 성공적으로 완료되었습니다!")
```

```python
import zipfile

# [파일 1] 가구별 소비액 예측 파일
consumption_df = test_features[['survey_id', 'hhid']].copy()
consumption_df['cons_ppp17'] = final_predictions
consumption_df.to_csv('predicted_household_consumption_2.csv', index=False)

# [파일 2] 설문 그룹별 빈곤 분포(CDF) 파일
thresholds = [3.17, 3.94, 4.60, 5.26, 5.88, 6.47, 7.06, 7.70, 8.40, 9.13,
              9.87, 10.70, 11.62, 12.69, 14.03, 15.64, 17.76, 20.99, 27.37]

def calculate_cdf(group):
    res = {'survey_id': group['survey_id'].iloc[0]}
    for t in thresholds:
        res[f'pct_hh_below_{t:.2f}'] = (group['cons_ppp17'] < t).mean()
    return pd.Series(res)

distribution_df = consumption_df.groupby('survey_id').apply(calculate_cdf).reset_index(drop=True)
distribution_df.to_csv('predicted_poverty_distribution_2.csv', index=False)

# [최종] 압축
with zipfile.ZipFile('final_submission_2.zip', 'w') as zf:
    zf.write('predicted_household_consumption_2.csv')
    zf.write('predicted_poverty_distribution_2.csv')

print("\n✅ 오류 수정 완료 및 'final_submission.zip' 생성 성공!")
```
