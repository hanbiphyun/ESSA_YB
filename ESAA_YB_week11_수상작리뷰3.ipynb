### 캐글 수상작 리뷰 3
##Playground Series
- 회귀 문제
- 주제 : 주어진 백팩의 특징을 기반으로 가격 예측하는 모델 만들기
- 특징 : 데이터 크기가 안 큼, 범주형 + 수치형 섞임, 피처 엔지니어링이 성능 차이를 만듬

**데이터 구성**
- Brand
- Material - 제품재질
- Size
- Compartments - 수납 공간 개수
- laptop compartment - 노트북 수납 공간 유무
- Waterproof
- Style - 캐주얼, 등산, 여행 등
- Color
- Weight capacity - 최대 적재 무게
- Price - 예측 목표

**변수 특성**
- 범주형 변수가 많음 (인코딩 전략)
- 이항 변수 존재
- 타깃 변수는 비선형성이 있을 것

==> 범주형 중심이기에 Tree 기반 모델(CatBoost)이 가장 유리할 것

**수상작 리뷰**

핵심 포인트
1. 범주형 변수를 target encoding 함 : 카테고리별 평균가격을 학습해서 가격과의 관계를 직접 반영함
2. 범주형 변수 조합 피처 생성
- brand+material
- style+size
- color + brand

==> 특정브랜드와 재질을 조합한 것이 가격을 올릴 수 있다는 인사이트 도출

3. CatBoost 단일 모델 사용
- 원-핫, 타깃 인코딩 자동 지원
- 범주형 많을수록 강함

==> 브랜드/스타일은 price에 직접적 영향을 줌 : 타깃인코딩의 영향이 컸음

**인사이트**
- 제품 가격은 기능보다 브랜드이미지의 영향이 크다
- 피처 엔지니어링이 회귀 문제에서도 결정적 역할을 한다
- 범주형 변수는 처리 전략을 잘 세워야하며 타깃인코딩, 순위 인코딩, 빈도 인코딩 등의 방법이 있다.

**보완할 점**
- 피처 생성 시 수작업으로 진행 => 조합 피처를 자동 생성하는 Featuretools, AutoFE 방법 시도해볼수있음.
- 많은 피처를 생성하여 어떤 피처가 핵심역할을 했는지 분석하기 어렵다 -> 상관관계, 변수 중요도를 더 파악하면 좋을듯
- 앙상블 접근이 아예 없었기 때문에 예측 안정화를 위해 CatBoost+ XGBoost+ LightGBM 스태킹의 유용성 고려해볼수있음
