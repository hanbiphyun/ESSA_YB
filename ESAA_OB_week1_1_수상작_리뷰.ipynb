주제 : 우주선 타이타닉호가 차원 이상 현상에 휘말렸는데, 승객들의 개인정보를 바탕으로 해당 승객이 다른 차원으로 이송되었는지 여부를 맞추는 이진 분류 문제

- 주요 쟁점 : 결측치 처리, 피처 엔지니어링 (피처가 갖고 있는 정보를 어떻게 추출하는지), 모델 선택 (정형데이터 분류 문제이므로 의사결정 나무 기반의 앙상블이 효과적)

- 위 수상작의 특징
  - TensorFlow Decision Forests 활용
  - 자동 전처리 : TFDF의 장점으로 모델이 알아서 처리해줌
- 배울 점 : 파이프라인 사용, 모델 시각화

```python
# 주요 코드
# 데이터 전처리

# Cabin (B/0/P) 형태를 Deck, Num, Side로 분리
dataset_df[["Deck", "Cabin_num", "Side"]] = dataset_df["Cabin"].str.split("/", expand=True)

# 필요 없는 원본 열 삭제
dataset_df = dataset_df.drop('Cabin', axis=1)
```

```python
# TFDF 모델 생성 및 학습

import tensorflow_decision_forests as tfdf

# 랜덤 포레스트 모델 생성 (이진 분류용)
rf_model = tfdf.keras.RandomForestModel(task=tfdf.keras.Task.CLASSIFICATION)

# 모델 학습 (pandas 데이터프레임을 tf dataset으로 변환 후 학습)
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="Transported")
rf_model.fit(train_ds)
```

```python
# 모델 성능 평가 (OOB 이용)

# Out-of-bag(OOB) 평가 지표 확인
logs = rf_model.make_inspector().evaluation()
print(f"Accuracy: {logs.accuracy:.4f}")
```
